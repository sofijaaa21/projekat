```python
import pandas as pd
from chembl_webresource_client.new_client import new_client
```

```python
# Target search
target = new_client.target
target_query = target.search('17β-HSD2')
targets = pd.DataFrame.from_dict(target_query)
targets
```

```python
selected_target = targets.target_chembl_id[2]
selected_target
```

```python
activity = new_client.activity
res = activity.filter(target_chembl_id=selected_target).filter(standard_type="IC50")
res
```

```python
df = pd.DataFrame.from_dict(res)
```

```python
df.head(10)
```

```python
df.to_csv('bioactivity_data.csv', index=False)
```

```python
df2 = df[df.standard_value.notna()]
df2
```

```python
bioactivity_class = []
for i in df2.standard_value:
  if float(i) >= 10000:
    bioactivity_class.append("inactive")
  elif float(i) <= 1000:
    bioactivity_class.append("active")
  else:
    bioactivity_class.append("intermediate")
```

```python
mol_cid = []
for i in df2.molecule_chembl_id:
  mol_cid.append(i)
```

```python
canonical_smiles = []
for i in df2.canonical_smiles:
  canonical_smiles.append(i)
```

```python
standard_value = []
for i in df2.standard_value:
  standard_value.append(i)
```

```python
data_tuples = list(zip(mol_cid, canonical_smiles, bioactivity_class, standard_value))
df3 = pd.DataFrame( data_tuples,  columns=['molecule_chembl_id', 'canonical_smiles', 'bioactivity_class', 'standard_value'])
```

```python
df3
```

```python
df3.to_csv('bioactivity_preprocessed_data.csv', index=False)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Učitavanje podataka iz CSV fajla
data = pd.read_csv('bioactivity_data.csv')

# Filtriranje potrebnih kolona i konverzija IC50 iz nM u µM
data['standard_value'] = data['standard_value'].astype(float)  # Osiguraj da su vrednosti brojevi
data['IC50_uM'] = data['standard_value'] / 1000  # Konvertuj nM u µM

# Inhibicija je obično u procentima, ako nemaš, možeš dodati stubac sa fiktivnim vrednostima
data['inhibition'] = data['pchembl_value']  # Ili koristi pravi stubac za inhibiciju

# Plotovanje
plt.figure(figsize=(10, 6))
plt.scatter(data['IC50_uM'], data['inhibition'], color='blue', alpha=0.7)
plt.title('Inhibicija u odnosu na IC50 koncentraciju')
plt.xlabel('IC50 (µM)')
plt.ylabel('Inhibicija (%)')
plt.grid()
plt.xscale('log')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Učitavanje podataka iz CSV fajla
data = pd.read_csv('bioactivity_data.csv')

# Filtriranje potrebnih kolona i konverzija IC50 iz nM u µM
data['standard_value'] = data['standard_value'].astype(float)  # Osiguraj da su vrednosti brojevi
data['IC50_uM'] = data['standard_value'] / 1000  # Konvertuj nM u µM

# Izračunavanje pIC50
data['pIC50'] = -np.log10(data['IC50_uM'])

# Inhibicija je obično u procentima, ako nemaš, možeš dodati stubac sa fiktivnim vrednostima
data['inhibition'] = data['pchembl_value']  # Ili koristi pravi stubac za inhibiciju

# Plotovanje
plt.figure(figsize=(10, 6))
plt.scatter(data['pIC50'], data['inhibition'], color='green', alpha=0.7)
plt.title('Inhibicija u odnosu na pIC50 koncentraciju')
plt.xlabel('pIC50')
plt.ylabel('Inhibicija (%)')
plt.grid()
plt.show()
```

```python
!pip install RdKit
import numpy as np
from rdkit import Chem
from rdkit.Chem import Descriptors, Lipinski
```

```python
# Inspired by: https://codeocean.com/explore/capsules?query=tag:data-curation

def lipinski(smiles, verbose=False):

    moldata= []
    for elem in smiles:
        mol=Chem.MolFromSmiles(elem)
        moldata.append(mol)

    baseData= np.arange(1,1)
    i=0
    for mol in moldata:

        desc_MolWt = Descriptors.MolWt(mol)
        desc_MolLogP = Descriptors.MolLogP(mol)
        desc_NumHDonors = Lipinski.NumHDonors(mol)
        desc_NumHAcceptors = Lipinski.NumHAcceptors(mol)

        row = np.array([desc_MolWt,
                        desc_MolLogP,
                        desc_NumHDonors,
                        desc_NumHAcceptors])

        if(i==0):
            baseData=row
        else:
            baseData=np.vstack([baseData, row])
        i=i+1

    columnNames=["MW","LogP","NumHDonors","NumHAcceptors"]
    descriptors = pd.DataFrame(data=baseData,columns=columnNames)

    return descriptors
```

```python
df_lipinski = lipinski(df.canonical_smiles)
```
